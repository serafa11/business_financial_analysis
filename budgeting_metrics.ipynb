```python
import pandas as pd
import numpy as np
from numpy_financial import npv, irr
```

```python
data = {"Period":[0, 1, 2, 3, 4],
      "Cash Flow": [-100000, 20000, 25000, 30000, 35000]}

df = pd.DataFrame(data)
```

```python
# for automatization of this first adjust df and rate
```

```python
rate = 0.1
```

```python
cashflows = df["Cash Flow"].to_list()
```

```python
npv_result = npv(rate,cashflows)
irr_result = irr(cashflows)
initial_investment = cashflows[0]
payback_period = 0
commulative_cashflow = 0
for period, cashflows in enumerate(cashflows):
  commulative_cashflow += cashflows
  if commulative_cashflow >= initial_investment:
    payback_period = period + 1
    break
```

Calculate PI

```python
pi = npv_result/abs(initial_investment)
```

```python
# Printing out results
print(f"NPV: ${npv_result:.2f}")
print(f"IRR: {irr_result:.2%}")
print(f"Payback Period: {payback_period} years")
print(f"PI: ${pi:.2f}")
```
